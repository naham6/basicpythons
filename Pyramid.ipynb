```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
fig, ax = plt.subplots(figsize=(8,8))
circle = plt.Circle((0.5, 0.5), 0.47, color='skyblue', ec='black', lw=1)
ax.add_artist(circle)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

circle = plt.Circle((0.5, 0.5), 0.47, color='skyblue', ec='black', lw=1)
ax.add_artist(circle)

ppoints = np.array([[0.5, 0.7],[0.15, 0.2],[0.85, 0.2]])
pyramid = Polygon(ppoints, color='darkorange', ec='black', lw=1)
ax.add_artist(pyramid)

shade_points = np.array([[0.5, 0.7],[0.15, 0.2],[0.32, 0.2]])
shade = Polygon(shade_points, color='brown', ec='black', lw=1, alpha=0.5)
ax.add_artist(shade)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
#ax.axis('off')

plt.show()
```
