```python
import networkx as nx
import matplotlib.pyplot as plt

# Assuming topics is a dictionary with nodes and their attributes
topics = {
    "Basic pthon": {"difficulty": 1},
    "ds": {"difficulty": 2},
    "advanced python": {"difficulty": 3},
    "ml": {"difficulty": 4}
}

# Initialize the graph
G = nx.Graph()

# Add nodes with attributes
for node, attrs in topics.items():
    G.add_node(node, **attrs)

# Edge attributes
edge_attributes = {
    ("Basic pthon", "ds"): {"w": 1, "prerequisite": True},
    ("ds", "advanced python"): {"w": 1.5, "prerequisite": True},
    ("advanced python", "ml"): {"w": 2, "prerequisite": True}
}

# Add edges with attributes
for edge, attrs in edge_attributes.items():
    G.add_edge(edge[0], edge[1], **attrs)

# Node positions
pos = nx.spring_layout(G)

# Draw the graph
nx.draw(G, pos, with_labels=True)

# Add node labels for difficulty
node_labels = {node: data["difficulty"] for node, data in G.nodes(data=True)}
for node, (x, y) in pos.items():
    plt.text(x, y - 0.05, node_labels[node], fontsize=10)

# Add edge labels for weight and prerequisite
edge_labels = {(u, v): f"w: {d['w']}, prereq :{d['prerequisite']}" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

# Display the plot
plt.show()
```

```python
print(f"No of nodes: {G.number_of_nodes()}")
print(f"No of edges: {G.number_of_edges()}")
```

```python
print(f"neighbour of basic python: {list(G.neighbors('Basic pthon'))}")
```
