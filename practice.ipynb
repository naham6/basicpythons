```python
movies={
    'The Shawshank Redemption':['drama'],
    'The Godfather':['crime','drama'],
    'The Dark Night':['action','crime'],
    'Fight Club':['drama','thriller']

}

user_genres =['action','thriller']
newlis=[]
for i,j in movies.items():
  for k in j:
    for l in user_genres :
      if k==l:
        newlis.append(i)
print(f'Recommendations = {newlis}')
```
