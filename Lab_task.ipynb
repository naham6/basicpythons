```python
# 2.
word = input("Enter a word: ")

def delete_1st_character(word):
    if len(word) > 1:
        return word[1:]
    else:
        return ""

re = delete_1st_character(word)
print("After deleting first character:", re)

```

```python
# 3.
lst = [1,0,8,10,12,3]
print(lst)

min_index = lst.index(min(lst))
max_index = lst.index(max(lst))

lst[min_index], lst[max_index] = lst[max_index], lst[min_index]

print(lst)
```

```python
#6.
def swap_word(words):
    word_li = words.split()
    if len(word_li) < 2:
        return "Input should contain at least two words"
    else:
        word1, word2 = word_li[:2]
        word_li[0], word_li[1] = word2, word1
        return ' '.join(word_li)


words = "Hey, Meera!"
print(swap_word(words))
```
